```python
file = 'Stone_Age'
with open('{}{}'.format(path, file), 'r', encoding='utf-8') as data:
    soup = BeautifulSoup(data, "lxml")
```

```python
body = soup.find(id='bodyContent')
```

```python
imgs = 0
for sibling in body.find_all('img'):
    try:
        if int(sibling['width']) >= 200:
            imgs += 1
            
    except KeyError:
        continue
```

```python
print(imgs)
```

```python
headers = 0
for header in body.find_all(['h1', 'h2', 'h3', 'h4', 'h5']):
    if header.text[:1] in ['E','T', 'C']:
        headers += 1
```

```python
print(headers)
```

```python
link_siblings_ = []
linkslen = 0
for link in body.find_all('a'):
    link_siblings = link.find_next_siblings()
    if len(link_siblings):
        linkslen_ = 1
        for link in link_siblings:
            if link.name == 'a': 
                linkslen_ += 1
                linkslen = max(linkslen_, linkslen)
            else: 
                linkslen_ = 0
        
        
        link_siblings_.append((linkslen_, link_siblings))

```

```python
sorted(link_siblings_, key=lambda link_sibling: link_sibling[0], reverse=True)
```

```python
print(linkslen)
```

```python
lists = 0
for tag in body.find_all(['ol', 'ul']):
    for parent in tag.parents:
        if parent.name in ['ol', 'ul']:
            break
    else:
        lists += 1
            
```

```python
print(lists)
```
